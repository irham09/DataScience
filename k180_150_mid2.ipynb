# K means Clustering

```python
colo = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi',
                                         'pedigree', 'age', 'label']
df2= pd.read_csv('pima.csv',names= colo)
df2= pd.DataFrame(df2)
display(df2)
```

```python
y=np.asarray(df2[['label']])
X=df2[['pregnant', 'insulin', 'bmi', 'skin', 'age','glucose','bp','pedigree']]
```

```python
from sklearn.cluster import KMeans
```

```python
def K_cluster(clt):
    kmeans = KMeans(n_clusters=clt, init='k-means++', n_init=10, max_iter=300, tol=0.0001, 
           verbose=0, random_state=None, copy_x=True, algorithm='auto')
    kmeans.fit(X)

    km_labels = kmeans.labels_
    #print(km_labels)
    cm_kmean = confusion_matrix(km_labels,y)
    print(cm_kmean)
```

```python
for ctr in range(9):
    K_cluster(ctr+1)
    
    
    
```
